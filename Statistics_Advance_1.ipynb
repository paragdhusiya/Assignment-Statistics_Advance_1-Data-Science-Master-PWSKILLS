Q1. What is the Probability density function?

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Generate random data from a normal distribution
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Create a range of values for the x-axis
x_values = np.linspace(-3, 3, 1000)

# Calculate the PDF for each x value using a normal distribution
pdf_values = norm.pdf(x_values, loc=np.mean(data), scale=np.std(data))

# Plot the histogram of the data
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue', label='Histogram')

# Plot the PDF curve
plt.plot(x_values, pdf_values, color='red', linewidth=2, label='PDF')

# Set labels and legend
plt.xlabel('Values')
plt.ylabel('Density')
plt.legend()

# Display the plot
plt.title('Probability Density Function (PDF)')
plt.show()
```

Q2. What are the types of Probability distribution?

```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, uniform, poisson

# Set seed for reproducibility
np.random.seed(42)

# Generate random data for each distribution
normal_data = np.random.normal(loc=0, scale=1, size=1000)
exponential_data = np.random.exponential(scale=1, size=1000)
uniform_data = np.random.uniform(low=0, high=1, size=1000)
poisson_data = np.random.poisson(lam=5, size=1000)

# Plot histograms and PDFs for each distribution
def plot_distribution(data, distribution_name, color):
    plt.hist(data, bins=30, density=True, alpha=0.7, color=color, label=f'{distribution_name} Data')
    plt.title(f'{distribution_name} Distribution')
    plt.xlabel('Values')
    plt.ylabel('Density')
    plt.legend()
    plt.show()

# Plot Normal Distribution
plot_distribution(normal_data, 'Normal', 'blue')

# Plot Exponential Distribution
plot_distribution(exponential_data, 'Exponential', 'green')

# Plot Uniform Distribution
plot_distribution(uniform_data, 'Uniform', 'orange')

# Plot Poisson Distribution
plot_distribution(poisson_data, 'Poisson', 'purple')
```

Q3. Write a Python function to calculate the probability density function of a normal distribution with
given mean and standard deviation at a given point.

```python
# Import necessary libraries
from scipy.stats import norm

def normal_pdf(x, mean, std_dev):
    """
    Calculate the probability density function (PDF) of a normal distribution at a given point.

    Parameters:
    - x: Point at which to calculate the PDF.
    - mean: Mean of the normal distribution.
    - std_dev: Standard deviation of the normal distribution.

    Returns:
    - pdf_value: Probability density function value at the given point.
    """
    pdf_value = norm.pdf(x, loc=mean, scale=std_dev)
    return pdf_value

# Example usage
mean_value = 0
std_dev_value = 1
point = 1.5

pdf_result = normal_pdf(point, mean_value, std_dev_value)
print(f"The PDF of a normal distribution at {point} is: {pdf_result}")
```

Q4. What are the properties of Binomial distribution? Give two examples of events where binomial
distribution can be applied.


```python
# Import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Function to plot binomial distribution
def plot_binomial_distribution(n, p, example_name):
    x_values = np.arange(0, n+1)
    pmf_values = binom.pmf(x_values, n, p)

    markerline, stemlines, baseline = plt.stem(x_values, pmf_values, basefmt=" ", label='Binomial PMF')

    # Set the color of the stems
    plt.setp(stemlines, 'color', 'blue')

    plt.title(f'Binomial Distribution - {example_name}')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.legend()
    plt.show()

# Example 1: Coin Flipping
n_coin_flipping = 3
p_coin_flipping = 0.5
example_name_coin_flipping = 'Coin Flipping (3 trials, fair coin)'

plot_binomial_distribution(n_coin_flipping, p_coin_flipping, example_name_coin_flipping)

# Example 2: Quality Control in Manufacturing
n_quality_control = 100
p_defective_item = 0.02
example_name_quality_control = 'Quality Control in Manufacturing (100 items)'

plot_binomial_distribution(n_quality_control, p_defective_item, example_name_quality_control)
```

Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution
with given mean at a given point.

```python
from scipy.stats import poisson

def poisson_cdf(x, mean):
    """
    Calculate the cumulative distribution function (CDF) of a Poisson distribution at a given point.

    Parameters:
    - x: Point at which to calculate the CDF.
    - mean: Mean of the Poisson distribution.

    Returns:
    - cdf_value: Cumulative distribution function value at the given point.
    """
    cdf_value = poisson.cdf(x, mu=mean)
    return cdf_value

# Example usage
mean_value = 3
point = 2

cdf_result = poisson_cdf(point, mean_value)
print(f"The CDF of a Poisson distribution with mean {mean_value} at {point} is: {cdf_result}")
```

Q7. How Binomial distribution different from Poisson distribution?


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson

# Function to plot Binomial distribution
def plot_binomial_distribution(n, p, example_name):
    x_values = np.arange(0, n+1)
    pmf_values = binom.pmf(x_values, n, p)

    markerline, stemlines, baseline = plt.stem(x_values, pmf_values, basefmt=" ", label='Binomial PMF')

    # Set the color of the stems
    plt.setp(stemlines, 'color', 'blue')

    plt.title(f'Binomial Distribution - {example_name}')
    plt.xlabel('Number of Successes')
    plt.ylabel('Probability')
    plt.legend()
    plt.show()

# Function to plot Poisson distribution
def plot_poisson_distribution(lam, example_name):
    x_values = np.arange(0, 15)
    pmf_values = poisson.pmf(x_values, lam)

    markerline, stemlines, baseline = plt.stem(x_values, pmf_values, basefmt=" ", label='Poisson PMF')

    # Set the color of the stems
    plt.setp(stemlines, 'color', 'green')

    plt.title(f'Poisson Distribution - {example_name}')
    plt.xlabel('Number of Events')
    plt.ylabel('Probability')
    plt.legend()
    plt.show()

# Example 1: Binomial Distribution (Coin Flipping)
n_binomial = 5
p_binomial = 0.5
example_name_binomial = 'Binomial (5 trials, fair coin)'

plot_binomial_distribution(n_binomial, p_binomial, example_name_binomial)

# Example 2: Poisson Distribution (Call Center)
lam_poisson = 2
example_name_poisson = 'Poisson (2 calls per minute)'

plot_poisson_distribution(lam_poisson, example_name_poisson)
```

Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the
sample mean and variance.

```python
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Generate a random sample from a Poisson distribution
sample_size = 1000
mean_poisson = 5
random_sample = np.random.poisson(lam=mean_poisson, size=sample_size)

# Calculate sample mean and variance
sample_mean = np.mean(random_sample)
sample_variance = np.var(random_sample, ddof=1)  # Using ddof=1 for sample variance

# Display the results
print(f"Sample Mean: {sample_mean}")
print(f"Sample Variance: {sample_variance}")
```

Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

```python
import numpy as np
from scipy.stats import binom, poisson

# Function to calculate mean and variance for Binomial distribution
def binomial_mean_variance(n, p):
    mean = n * p
    variance = n * p * (1 - p)
    return mean, variance

# Function to calculate mean and variance for Poisson distribution
def poisson_mean_variance(lam):
    mean = lam
    variance = lam
    return mean, variance

# Example 1: Binomial Distribution
n_binomial = 10
p_binomial = 0.3

binomial_mean, binomial_variance = binomial_mean_variance(n_binomial, p_binomial)
print(f"Binomial Distribution - Mean: {binomial_mean}, Variance: {binomial_variance}")

# Example 2: Poisson Distribution
lam_poisson = 3

poisson_mean, poisson_variance = poisson_mean_variance(lam_poisson)
print(f"Poisson Distribution - Mean: {poisson_mean}, Variance: {poisson_variance}")
```

Q10. In normal distribution with respect to mean position, where does the least frequent data appear?

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Function to plot normal distribution
def plot_normal_distribution(mean, std_dev):
    x_values = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 1000)
    pdf_values = norm.pdf(x_values, mean, std_dev)

    plt.plot(x_values, pdf_values, label='Normal Distribution')
    plt.title('Normal Distribution')
    plt.xlabel('X')
    plt.ylabel('Probability Density Function (PDF)')
    plt.legend()
    plt.show()

# Example: Normal Distribution
mean_normal = 0
std_dev_normal = 1

# Plotting the normal distribution
plot_normal_distribution(mean_normal, std_dev_normal)
```
